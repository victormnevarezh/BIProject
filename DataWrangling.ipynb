```python
import numpy as np
import pandas as pd  
```

```python
# Specify the file path
file_path = r'U:\BIProject\vgsales.csv'

# Load CSV file from file path
df = pd.read_csv(file_path)

df.head(0)
```

```python
# Remove rows with missing values
df = df.dropna()
```

```python
# Convert 'Year' column from float to int
df[['Year',
    'NA_Sales', 
    'EU_Sales', 
    'JP_Sales', 
    'Other_Sales', 
    'Global_Sales']] = df[['Year',
                        'NA_Sales', 
                        'EU_Sales', 
                        'JP_Sales', 
                        'Other_Sales', 
                        'Global_Sales']].astype(int)
```

```python
df[['NA_Sales', 
    'EU_Sales', 
    'JP_Sales', 
    'Other_Sales', 
    'Global_Sales']] = df[['NA_Sales', 
                           'EU_Sales', 
                           'JP_Sales', 
                           'Other_Sales', 
                           'Global_Sales']] * 1000000
```

```python
df.to_csv('vgsalesCleaned.csv', index=False)
```
