```python
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('uncleaned_dataset.csv')

# Print the first few rows
# print(data.head())

# Get the column names
column_names = data.columns
#column_names

len(data)
```

```python
data.dropna(inplace=True)
len(data)
```

```python
selected_columns = ['Artist', 
                    'Track', 
                    'Album', 
                    'Album_type', 
                    'Danceability', 
                    'Energy', 
                    'Key', 
                    'Loudness', 
                    'Speechiness', 
                    'Acousticness', 
                    'Instrumentalness', 
                    'Liveness', 
                    'Valence', 
                    'Tempo', 
                    'Duration_ms',
                    'Stream',
                    'Views', 
                    'Likes', 
                    'Comments', 
                    'official_video']
subsetDeletedColumns = data[selected_columns]
subsetDeletedColumns.head()
```

```python
renameColumns = {'Stream': 'SpotifyStream',
                 'Views': 'YoutubeViews',
                 'official_video': 'OfficialVideo'
                }
mainDataset = subsetDeletedColumns.rename(columns=renameColumns)
mainDataset.head()
```

```python
# Filter for values not equal to "album" or "single" in the "Album_type" column
filtered_data = mainDataset[~mainDataset['Album_type'].isin(['album', 'single', 'compilation'])]

# Print the filtered data
print(filtered_data['Album_type'])
```

```python
# Replace values in the "Album_type" column
replacement_mapping = {'single': 1, 'album': 2, 'compilation': 3}
mainDataset['Album_type'] = mainDataset['Album_type'].replace(replacement_mapping)
mainDataset.head()
```

```python
# Define the columns to find the range for
columns_to_check = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
                    'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']

# Find the range of values for each column
for column_name in columns_to_check:
    column_min = mainDataset[column_name].min()
    column_max = mainDataset[column_name].max()
    print(f"The range of values in the {column_name} column is from {column_min} to {column_max}.")
```

```python
# Define the columns to check
columns_to_check = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
                    'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']

# Filter values below 0 for each column
for column_name in columns_to_check:
    values_below_zero = mainDataset[mainDataset[column_name] < 0][column_name]
    print(f"Values below 0 in {column_name}:")
    print(values_below_zero)
    print()
```

```python
# Define the column name
column_name = 'OfficialVideo'
# Replace values in the column
mainDataset[column_name].replace({True: 1, False: 0}, inplace=True)
mainDataset
```

```python
# Assuming 'mainDataset' is your DataFrame

# Specify the columns that may contain numbers in scientific notation
columns_to_convert = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
                      'Acousticness', 'Instrumentalness', 'Liveness', 'Valence',
                      'Tempo', 'Duration_ms', 'YoutubeViews', 'SpotifyStream', 'Likes', 'Comments']

# Convert scientific notation to decimal format for the specified columns
mainDataset[columns_to_convert] = mainDataset[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```
