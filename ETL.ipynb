```python
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('uncleaned_dataset.csv')

# Print the first few rows
# print(data.head())

# Get the column names
column_names = data.columns
#column_names

len(data)
```

```python
data.dropna(inplace=True)
len(data)
```

```python
selected_columns = ['Artist', 
                    'Track', 
                    'Album', 
                    'Album_type', 
                    'Danceability', 
                    'Energy', 
                    'Key', 
                    'Loudness', 
                    'Speechiness', 
                    'Acousticness', 
                    'Instrumentalness', 
                    'Liveness', 
                    'Valence', 
                    'Tempo', 
                    'Duration_ms',
                    'Stream',
                    'Views', 
                    'Likes', 
                    'Comments', 
                    'official_video']
subsetDeletedColumns = data[selected_columns]
subsetDeletedColumns.head()
```

```python
renameColumns = {'Stream': 'SpotifyStream',
                 'Views': 'YoutubeViews',
                 'official_video': 'OfficialVideo'
                }
mainDataset = subsetDeletedColumns.rename(columns=renameColumns)
mainDataset.head()
```

```python
# Filter for values not equal to "album" or "single" in the "Album_type" column
filtered_data = mainDataset[~mainDataset['Album_type'].isin(['album', 'single', 'compilation'])]

# Print the filtered data
print(filtered_data['Album_type'])
```

```python
# Replace values in the "Album_type" column
replacement_mapping = {'single': 1, 'album': 2, 'compilation': 3}
mainDataset['Album_type'] = mainDataset['Album_type'].replace(replacement_mapping)
mainDataset.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Define the selected columns
selected_columns = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness',
                    'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms']

# Prepare the data for clustering
selected_data = clustered_data[selected_columns].values
spotify_stream = clustered_data['SpotifyStream'].values.reshape(-1, 1)

# Perform data preprocessing (scaling)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data)

# Perform K-means clustering for each column
num_clusters = 3  # Number of clusters
fig, axs = plt.subplots(4, 3, figsize=(15, 15))  # Create subplots

for i, column in enumerate(selected_columns):
    # Create a K-means model
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)

    # Plot the clusters
    row = i // 3
    col = i % 3
    axs[row, col].scatter(selected_data[:, i], spotify_stream, c=cluster_labels, cmap='viridis')
    axs[row, col].set_xlabel('SpotifyStream')
    axs[row, col].set_ylabel(column)

plt.tight_layout()
plt.show()
```

```python
def plot_cluster(column_data, target_data, num_clusters):
    # Create a K-means model
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(column_data)

    # Plot the clusters
    plt.scatter(column_data, target_data, c=cluster_labels, cmap='viridis')
    plt.xlabel('Column Data')
    plt.ylabel('SpotifyStream')
    plt.title(f'Clusters of {column_data.name} vs SpotifyStream')
    plt.show()
```
